```python
import sys
sys.path.append("../../")  # To access teleport_dynamic module
from teleport_dynamic.analytical import calculate_stream_values
```

```python

# Parameters for minus1_2, 4 agents
gamma = 0.99
num_agents = 4
p_hit = 5.0 / (9.0 * 9.0)  # 5 apples on 9x9 grid

# 1. Calculate for Self (Picker = -1.0)
rho_self = -1.0
v_rand_s, v_off_s, v_on_s = calculate_stream_values(rho_self, gamma, num_agents, p_hit)
print(f"Self Streams (rho=-1): RAND={v_rand_s:.4f}, OFF={v_off_s:.4f}, ON={v_on_s:.4f}")

# 2. Calculate for Others (Other = 2/3)
rho_other = 2.0 / 3.0
v_rand_o, v_off_o, v_on_o = calculate_stream_values(rho_other, gamma, num_agents, p_hit)
print(f"Other Streams (rho=0.66): RAND={v_rand_o:.4f}, OFF={v_off_o:.4f}, ON={v_on_o:.4f}")

# 3. Approximate Total Value (Self on Apple + 3 Others on Apples)
# This estimates the value target the network is trying to learn
approx_val = v_rand_s + 3 * v_on_o
print(f"Approximate V(s) for Self_Picker: {approx_val:.4f}")
```
