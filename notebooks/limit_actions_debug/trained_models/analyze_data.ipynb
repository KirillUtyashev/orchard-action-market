```python
import sys
sys.path.append("../../../")
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python

# Load both DataFrames
df_values = pd.read_pickle("predicted_trajectory_values_6x6_agents2.pkl")
```

```python
print(df_values.keys())
```

```python
print(df_values["error"].mean())
print(df_values["centralized_team_value"].mean())
print(df_values["decentralized_team_value"].mean())
```

```python
# index 63944
problem_states = df_values.iloc[63940:63946]
for index, row in problem_states.iterrows():
    print(f"""On timestep {row['timestep']}, centralized_value: {row['centralized_team_value']}, 
          decentralized_team_value: {row['decentralized_team_value']}, error: {row['error']}""")
    print(row["state_object"])
```

```python
top_10_value_outliers = df_values.sort_values(by='error', ascending=False).head(10)
for index, row in top_10_value_outliers.iterrows():
    print(f"""On timestep {row['timestep']}, centralized_value: {row['centralized_team_value']}, 
          decentralized_team_value: {row['decentralized_team_value']}, error: {row['error']}""")
    print(row["state_object"])
```

```python
top_10_value_outliers_good = df_values.sort_values(by='error', ascending=True).head(10)
for index, row in top_10_value_outliers_good.iterrows():
    print(f"""On timestep {row['timestep']}, centralized_value: {row['centralized_team_value']}, 
          decentralized_team_value: {row['decentralized_team_value']}, error: {row['error']}""")
    print(row["state_object"])
```
